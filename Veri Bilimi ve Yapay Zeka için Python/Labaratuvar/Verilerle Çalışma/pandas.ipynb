```python
#DataFrame
data = {
    "Name": ["Ali", "Ayşe", "Mehmet", "Zeynep", "Erhan", "Mustafa"],
    "Age": [25, 30, 22, 27, 23, 27],
    "City": ["İstanbul", "Ankara", "İzmir", "Bursa", "Antalya", "Konya"]
}
df = pd.DataFrame(data)
df
```

```python
df["NameLength"] = df["Name"].apply(lambda x: len(x))
df
```

```python
df.loc[2, "Age"] = None #Boş değer oluşturur.
df.isnull().sum() #Kaç tane boş değer var
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean()) #Ortalama ile doldur
df
```

```python
#Bir DataFrame oluşturup csv'ye kaydetme
df.to_csv("students.csv", index = False)
```
