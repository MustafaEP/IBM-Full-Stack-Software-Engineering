```python
#Örnek uygulama öğrenci analizi
notlar = np.array([75, 82, 90, 67, 88, 92, 58])

print("Ortalama:", np.mean(notlar))
print("Maks:", np.max(notlar))
print("Başarılı (>80):", notlar[notlar > 80])
```
