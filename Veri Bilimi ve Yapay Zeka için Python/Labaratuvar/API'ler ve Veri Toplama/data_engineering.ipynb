```python
"""
loc() etiket (label) tabanlı veri seçme yöntemidir. 
Bu, seçmek istediğimiz satırın veya sütunun adını (etiketini) 
parametre olarak vermemiz gerektiği anlamına gelir.
"""

# İlk Satır
df.loc[0]
```

```python
import json
person = {
    'first_name' : 'Mark',
    'last_name' : 'abc',
    'age' : 27,
    'address': {
        "streetAddress": "21 2nd Street",
        "city": "New York",
        "state": "NY",
        "postalCode": "10021-3100"
    }
}
```

```python
with open('person.json', 'w') as f:  # writing JSON object
    json.dump(person, f)
```

```python
# Serializing json  
json_object = json.dumps(person, indent = 4) 
  
# Writing to sample.json 
with open("sample.json", "w") as outfile: 
    outfile.write(json_object) 
```

```python
print(json_object)
```

```python
import json 
  
# Opening JSON file 
with open('sample.json', 'r') as openfile: 
  
    # Reading from json file 
    json_object = json.load(openfile) 
  
print(json_object) 
print(type(json_object)) 
```
