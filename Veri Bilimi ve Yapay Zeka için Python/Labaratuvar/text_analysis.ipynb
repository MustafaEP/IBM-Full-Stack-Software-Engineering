```python
givenstring="Lorem ipsum dolor! diam amet, consetetur Lorem magna. sed diam nonumy eirmod tempor. diam et labore? et diam magna. et diam amet."
```

```python
class TextAnalyzer(object):

    def __init__(self, text):
        frm_text = text.replace('.', '').replace('!', '').replace('?', '').replace(',','')
        frm_text = frm_text.lower()
        self.frm_text = frm_text

    def freqAll(self):
        splited_text = self.frm_text.split(' ')
        word_freq = {}
        for word in set(splited_text):
            word_freq[word] = splited_text.count(word)
        return word_freq

    def freqOf(self, word):
        freqDict = self.freqAll()
        if word in freqDict:
            return freqDict[word]
        else:
            return 0
```

```python
analyz = TextAnalyzer(givenstring)
```

```python
print("Formatted Text:", analyz.frm_text)
```

```python
freqMap = analyz.freqAll()
print(freqMap)
```
