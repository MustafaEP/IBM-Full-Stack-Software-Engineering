```python
# First function example: Add 1 to a and store as b

def add(a):
    """
    add 1 to a
    """
    b = a + 1
    print(a, "if you add one", b)
    return(b)
```

```python
# Get a help on add function

help(add)
```

```python
# Call the function add()

add(1)
```
